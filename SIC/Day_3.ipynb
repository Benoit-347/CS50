```python
# tuple, these are immutable.
"""
====================  PYTHON TUPLE INTERNALS  ====================

CPython C struct (simplified):

    typedef struct {
        PyObject_VAR_HEAD        # type, refcount, size
        PyObject *ob_item[1];     # fixed-size array of object pointers
    } PyTupleObject;

Key points:

1. Tuple stores a *fixed-size* C array of pointers (ob_item[]).
2. These pointers are assigned only once at tuple creation.
3. Memory layout is contiguous and has NO extra capacity.
4. No C API exists to change ob_item[] or resize the tuple.
5. Therefore the container cannot be modified → tuple is immutable.

Immutability rule:
- Container is frozen (fixed pointer-array).
- Objects inside can still be mutable, but the tuple’s pointers cannot change.

==================================================================

"""
```

```python
# data structure of a list:
"""
=====================  PYTHON LIST INTERNALS  =====================

CPython C struct (simplified):

    typedef struct {
        PyObject_VAR_HEAD         # type, refcount, logical size
        PyObject **ob_item;       # pointer to a dynamic array of pointers
        Py_ssize_t allocated;     # actual allocated capacity (>= size)
    } PyListObject;

Key points:

1. list stores a *dynamic* array of object pointers (ob_item).
2. Memory is overallocated (allocated >= size) to support growth.
3. append(), insert(), pop(), etc. can mutate ob_item[] directly.
4. When capacity is exceeded, list reallocates a larger block
   and copies pointers (amortized O(1) append).
5. Because ob_item[] is writable + resizable → list is mutable.

Mutability rule:
- Container is flexible (dynamic pointer-array).
- Elements inside can be replaced, removed, appended, or inserted.
- List identity (id()) stays same unless reallocation occurs internally.

===================================================================

"""
```

```python
# list operations: 

"""
===================== PYTHON LISTS: CHEAT-SHEET (executable) =====================

Quick reference of standard & important list operations with short comments and examples.

Keep this file runnable: examples print results or show behavior. Tweak at will.

===============================================================================
"""

from copy import deepcopy

# -----------------------
# Creation / conversion
# -----------------------
L = [1, 2, 3]               # literal
T = (4, 5)
L_from_T = list(T)         # convert tuple -> list
print("L, L_from_T:", L, L_from_T)

# -----------------------
# Indexing & slicing
# -----------------------
print("L[0]:", L[0])        # indexing (0-based)
print("L[-1]:", L[-1])      # negative index
print("L[1:3]:", L[1:3])    # slice (new list)
print("L[:]:", L[:])        # shallow copy via slice

# -----------------------
# Basic mutating ops
# -----------------------
a = [1, 2]
print("id before append:", id(a))
a.append(3)                # append single element (in-place)
print("after append:", a, "id:", id(a))

a.extend([4, 5])          # extend with iterable (in-place)
print("after extend:", a)

a.insert(0, 0)            # insert at index
print("after insert:", a)

val = a.pop()             # pop last (or pop(i))
print("popped:", val, "list now:", a)

a.remove(2)               # remove first occurrence of value
print("after remove(2):", a)

a.clear()                 # remove all items
print("after clear:", a)

# -----------------------
# Non-mutating / queries
# -----------------------
L = [10, 20, 30, 20]
print("len(L):", len(L))
print("20 in L:", 20 in L)         # membership
print("count(20):", L.count(20))   # frequency
print("index(30):", L.index(30))   # first index of value

# safe index lookup
try:
    print("index(99):", L.index(99))
except ValueError:
    print("index(99) raised ValueError (not found)")

# -----------------------
# Sorting & reversing
# -----------------------
unsorted = [3, 1, 4, 2]
sorted_new = sorted(unsorted)      # returns a new list
print("sorted_new:", sorted_new, "orig:", unsorted)

unsorted.sort()                    # in-place sort
print("after in-place sort:", unsorted)

# sort with key & reverse
names = ["lee", "Ann", "bob"]
names.sort(key=str.lower, reverse=False)
print("names sorted case-insensitively:", names)

# reverse
names.reverse()                    # in-place reverse
print("reversed names:", names)

# -----------------------
# Shallow vs Deep copy
# -----------------------
outer = [[1, 2], [3, 4]]
shallow = list(outer)  # or outer[:] or outer.copy()
deep = deepcopy(outer)

outer[0].append(99)    # mutate inner object
print("outer:", outer)
print("shallow after inner mutate (shares inner lists):", shallow)
print("deep after inner mutate (independent):", deep)

# -----------------------
# Concatenation & repetition
# -----------------------
a = [1, 2]
b = [3, 4]
print("concat (+):", a + b)   # new list
a += b                        # in-place extend via +=
print("a after += b:", a)
print("repeat *:", [0] * 3)   # repetition

# -----------------------
# Iteration patterns
# -----------------------
for i, v in enumerate(["a", "b", "c"]):
    print("enumerate:", i, v)

# list comprehension (fast & pythonic)
squares = [x * x for x in range(6)]
print("squares:", squares)

# filter/map via comprehensions (preferred for readability)
even = [x for x in range(10) if x % 2 == 0]
print("even:", even)

# -----------------------
# Useful builtins that work well with lists
# -----------------------
nums = [1, 2, 3, 4]
print("sum:", sum(nums))
print("min/max:", min(nums), max(nums))
print("any(all):", any([0, 0, 3]), all([1, 2, 3]))

# join (convert list of strings to single string)
words = ["hello", "world"]
print("join:", " ".join(words))

# -----------------------
# Slicing operations as mutation (assign / delete)
# -----------------------
arr = [0, 1, 2, 3, 4]
arr[1:3] = [9, 9]       # replace slice (in-place)
print("slice replace:", arr)
arr[1:2] = []           # delete slice (in-place)
print("slice delete:", arr)

# -----------------------
# Copy semantics & identity notes
# -----------------------
orig = [1, 2, 3]
shallow_copy = orig     # alias, same object
slice_copy = orig[:]    # new list (shallow)
print("orig id:", id(orig), "shallow_copy id:", id(shallow_copy), "slice_copy id:", id(slice_copy))

# modify element in place
orig[0] = 999
print("after modifying orig:", orig, "slice_copy remains:", slice_copy)

# -----------------------
# Performance & mutability notes (short)
# -----------------------
# - append/pop from end: amortized O(1)
# - pop(0) / insert(0, x): O(n) (shifts elements)
# - random access by index: O(1)
# - slicing/copying: O(k) where k = length of slice
# - use list comprehensions / builtins / itertools for faster loops

# -----------------------
# Misc: converting iterables
# -----------------------
it = map(lambda x: x * 2, [1, 2, 3])
list_from_it = list(it)   # consume iterator into list
print("from map:", list_from_it)

# -----------------------
# Final quick examples to illustrate references (object vs binding)
# -----------------------
a = 10
parent = [a]
a = 50                  # rebind a -> new object; list still points to old 10
print("parent after rebind of a:", parent)

mut = [1, 2]
container = [mut]       # list stores reference to the inner list
mut.append(3)           # mutating the inner list shows inside container
print("container after mutating inner list:", container)

# ========================= End of cheat-sheet =========================
```

```python
# finding the shortest string form a list
input_list = ["apple", "banana", "kiwi", "cherry", "mango", "fig", "blueberry"] 

shortest_str = input_list[0]
len_shortest = len(shortest_str)

for ele in input_list:
    len_new = len(ele)
    if len_shortest > len_new:  # algo: first occurence, '>' last occurence, '>='
        len_shortest = len_new
        shortest_str = ele

print(f"The shortes string in ths list is: {shortest_str}, it has length: {len_shortest}")
```

```python
# find larget string in list:
input_list = ["apple", "banana", "kiwi", "cherry", "mango", "fig", "blueberry"]

longest_str = input_list[0]
len_longest = len(longest_str)

for ele in input_list:
    len_new = len(ele)
    if len_longest < len_new:
        len_longest = len_new
        longest_str = ele

print(f"The longest string in the list is: {longest_str}, it has length: {len_longest}")
```

```python
# a list in py of 5 elements initialized, does not have only 5 ref spaces, but also has extra space apart from the ref of initialized elements into list obj.
    # if The list is full. If the underlying array is full, Python performs a reallocation.

"""
=============================================================================
   PYTHON LIST OPERATIONS: MEMORY & PERFORMANCE (ASCII NOTES)
=============================================================================

1. CONCATENATION (+) vs EXTEND vs APPEND
+----------------+---------------------------+----------------------+--------+
|   Operation    |      Memory Action        |     Ref Handling     | Speed  |
+----------------+---------------------------+----------------------+--------+
| list1 + list2  | NEW List Object created   | Copies refs from     | O(N+M) |
|                | Alloc: len(L1) + len(L2)  | BOTH lists to new    | (Slow) |
+----------------+---------------------------+----------------------+--------+
| list1.extend() | SAME List Object (L1)     | Copies refs from     | O(M)   |
|                | Alloc: Resize if full     | L2 into L1 slots     | (Fast) |
+----------------+---------------------------+----------------------+--------+
| list1.append(a)| SAME List Object (L1)     | Copies ref of 'a'    | O(1)   |
|                | Alloc: Resize if full     | into next slot       | (Best) |
+----------------+---------------------------+----------------------+--------+

2. DEEP DIVE: list.append(a)
   [List Header] -> [Size: 5 | Capacity: 8]
                       |
   [C-Array Ptrs] -> [Ref1][Ref2][Ref3][Ref4][Ref5][EMPTY][EMPTY][EMPTY]
                                               ^
                                               |
   list.append(a) writes ADDRESS of 'a' here --+
   
   * NO NEW OBJECT: 'a' is NOT cloned/copied.
   * NO NEW LIST:   'list1' stays at same address (unless resize triggered).
   * COST:          Just writing one pointer (64-bit int).

3. RESIZING (Amortization)
   * Lists are "Dynamic Arrays" (over-allocated).
   * If [Capacity] is full:
     1. Python mallocs new block (~1.125x bigger).
     2. Copies old refs to new block.
     3. Frees old block.
"""
```

```python
# count fn (build in fn)
    # it searches the iterable for specified element and return the number of occurences.
        # similar to .remove(ele) but does not remove, but seaches and does not stop at first occurence, but searches all matches.
list_1 = [10, 20, 30, 40, 50, 30, 10, 20]
ele = 10
print(f"Number of occurence of the element {ele} is: ", end = '')
output = list_1.count(ele)
print(output)
# prints number of specified elements within the sequintial data obj.
```

```python
# using fast method: sum along with another fast method: list comprehension (he iteration pattern is optimized in C, avoiding the overhead of repeated append() function calls.)

def icount(iterable, target):
    return sum(1 for x in iterable if x == target)

# For ALL iterables (even those without .count)
print("Set icount:", icount({1,2,2,3}, 2))              # sets dedupe → 1 or 0
print("Dict-keys icount:", icount({2:10,3:20}, 2))
print("Generator icount:", icount((x % 3 for x in range(10)), 1))
print("range icount:", icount(range(10), 5))
print("Bytes icount:", icount(b"ababa", ord("a")))
```
