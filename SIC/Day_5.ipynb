```python
# fns like print, min take multiple arguments, they are called arbitrary functions
    # use *preix at parameter to show variable length arg fn
def greet(*names):      # arbitrary parameter
    for name in names:
        print("Hello", name)

greet("Benoit", "Anish")
```

```python
# default parameter
def greet_s(name, n = 1):   # def greet(*names, n = 1) is bugged as at call, even the 3rd arg will be taken as name, and 
    for _ in range(n):
        print("Hello", name)

greet_s("Benoit")
print()
greet_s("Benoit", 3)
```

```python
# caesar cipher
text = "records are in banglaore"
text_encrypted = ""
for i in text:                                              # obtain char start = 'a' - 97 = 0
                                                                # shift by 3 = (start + 3) %26
                                                                    # get char back from start = shifted + 97
    text_encrypted += chr( (ord(i) - 97 + 3)%26 + 97)   # encrypted char should be a lower case aplhabet, so chain => '%'

print(f"The encrypted text is: {text_encrypted}")
```

```python
# =============================================================================
#  BIRD'S EYE VIEW: PYTHON FUNCTIONS, METHODS, AND WHERE THEY LIVE
#  For Samsung Innovation (C&P) + CS50P style Python learning
#
#  Goal: stop thinking "random functions floating in air",
#        start thinking "which OBJECT / CLASS does this belong to?"
#
#  Use this as a quick map: what is a builtin, what is a str/list/dict method,
#  what comes from modules like math/random, etc.
# =============================================================================


# -----------------------------------------------------------------------------
# 0. TERMINOLOGY (VERY IMPORTANT)
# -----------------------------------------------------------------------------
# builtin function      → global fn, used like:    len(x), print(x)
# method on a class    → attached to type, used like:  s.lower(), lst.append(...)
# module function      → lives in module, used like:   math.sqrt(x), random.randint(a,b)
#
# Example:
# - len("hi")     → builtin
# - "hi".upper()  → str method
# - math.sqrt(9)  → function in math module
# -----------------------------------------------------------------------------


# -----------------------------------------------------------------------------
# 1. BUILT-IN FUNCTIONS YOU PROBABLY USED (NO DOT BEFORE THEM)
# -----------------------------------------------------------------------------
# I/O and conversion
# ------------------
# print(...)          → send text to stdout
# input(prompt)       → read a line of text from stdin
# int(x)              → convert to int
# float(x)            → convert to float
# str(x)              → convert to str
# bool(x)             → truthiness to bool

# Sequence / collection tools
# ---------------------------
# len(x)              → length of sequence/container
# range(start, stop, step=1) → arithmetic progression iterator
# list(iterable)      → build list
# tuple(iterable)     → build tuple
# dict(iterable/kwargs) → build dict
# set(iterable)       → build set
# sorted(iterable, key=None, reverse=False) → new sorted list
# sum(iterable, start=0)    → sum of numbers
# min(iterable, *args, key=None) → min element or value
# max(iterable, *args, key=None) → max element or value
# any(iterable)       → True if any element is truthy
# all(iterable)       → True if all elements are truthy
# reversed(seq)       → reverse iterator (needs sequence or __reversed__)

# Loop helpers and functionals
# ----------------------------
# enumerate(iterable, start=0) → yields (index, item)
# zip(*iterables)              → iterator of tuples (parallel iteration)
# map(func, iterable)          → apply func lazily
# filter(func, iterable)       → keep items where func(item) is True
# next(iterator, default?)     → get next item from iterator
# iter(obj)                    → get iterator from iterable

# Int / math-ish
# --------------
# abs(x)              → absolute value
# round(x, ndigits=0) → rounding
# pow(x, y, mod=None) → exponentiation with optional mod
# divmod(a, b)        → (a // b, a % b)

# Type / introspection
# --------------------
# type(x)             → returns the type object
# isinstance(obj, cls_or_tuple) → True if obj is instance of given type(s)
# id(obj)             → memory-identity-ish integer (implementation dependent)

# File / context manager
# ----------------------
# open(filename, mode="r", encoding="utf-8") → returns file object (see file methods below)


# -----------------------------------------------------------------------------
# 2. CLASS: str  (STRING METHODS)
# -----------------------------------------------------------------------------
# Strings are immutable text sequences. Methods always return NEW strings.

# Case operations
# ---------------
# s.lower()
# s.upper()
# s.title()       → "hello world" → "Hello World"
# s.capitalize()  → "hello" → "Hello"
# s.swapcase()

# Trimming and stripping
# ----------------------
# s.strip()       → remove whitespace both sides
# s.lstrip()
# s.rstrip()

# Searching
# ---------
# s.find(sub)         → index or -1
# s.rfind(sub)
# s.index(sub)        → like find, but raises ValueError if not found
# s.startswith(prefix)
# s.endswith(suffix)

# Replace / split / join
# ----------------------
# s.replace(old, new, count=-1)
# s.split(sep=None)           → split to list of substrings
# s.rsplit(sep=None)
# s.splitlines()
# sep.join(iterable_of_str)   → " ".join(["hello","world"])

# Classification (boolean tests)
# ------------------------------
# s.isalpha()
# s.isdigit()
# s.isalnum()
# s.isspace()
# s.islower()
# s.isupper()
# s.istitle()

# Formatting
# ----------
# "{} {}".format(a, b)
# f"{a} {b}"   → f-string (syntax feature, not a method)


# -----------------------------------------------------------------------------
# 3. CLASS: list  (LIST METHODS)
# -----------------------------------------------------------------------------
# Lists are mutable sequences (dynamic arrays).

# In-place modifiers
# ------------------
# lst.append(x)          → add at end
# lst.extend(iterable)   → append all items from iterable
# lst.insert(index, x)   → insert at position
# lst.remove(x)          → remove first occurrence of value, ValueError if not found
# lst.pop(index=-1)      → remove and return item at index (default last)
# lst.clear()            → remove all items
# lst.sort(key=None, reverse=False)  → in-place sort
# lst.reverse()          → in-place reverse

# Info / shallow copy
# -------------------
# lst.index(x)       → first index of value x, or ValueError
# lst.count(x)       → frequency of x
# lst.copy()         → shallow copy (new list, same element references)


# -----------------------------------------------------------------------------
# 4. CLASS: dict  (DICTIONARY METHODS)
# -----------------------------------------------------------------------------
# Dict is a hash map: key → value (keys must be hashable).

# Access / query
# --------------
# d.get(key, default=None)    → value or default (no KeyError)
# d.keys()                    → view of keys
# d.values()                  → view of values
# d.items()                   → view of (key, value) pairs

# Modification
# ------------
# d.update(other_mapping_or_pairs)   → merge/update keys
# d.pop(key, default=sentinel)      → remove key and return value; default if missing
# d.popitem()                       → remove and return an arbitrary (key, value) pair
# d.setdefault(key, default=None)   → like get, but also inserts default if missing
# d.clear()                         → remove all items

# Copy
# ----
# d.copy()         → shallow copy (new dict, same key/value refs)


# -----------------------------------------------------------------------------
# 5. CLASS: set  (SET METHODS)
# -----------------------------------------------------------------------------
# Set = unordered collection of unique hashable elements.

# Basic modification
# ------------------
# s.add(x)
# s.remove(x)          → KeyError if not present
# s.discard(x)         → no error if missing
# s.pop()              → remove & return an arbitrary element
# s.clear()

# Set algebra
# -----------
# s.union(t)           or: s | t
# s.intersection(t)    or: s & t
# s.difference(t)      or: s - t
# s.symmetric_difference(t)  or: s ^ t

# Relations
# ---------
# s.issubset(t)        or: s <= t
# s.issuperset(t)      or: s >= t
# s.isdisjoint(t)


# -----------------------------------------------------------------------------
# 6. CLASS: tuple  (TUPLE METHODS)
# -----------------------------------------------------------------------------
# Tuple = immutable sequence. Very few methods.

# t.count(x)    → how many times x appears
# t.index(x)    → index of first occurrence x (or ValueError)

# Everything else is via:
# - indexing (t[i])
# - slicing (t[a:b])
# - len(t), etc.  (builtins, not methods)


# -----------------------------------------------------------------------------
# 7. RANGE OBJECT (CLASS: range)
# -----------------------------------------------------------------------------
# range(start, stop, step=1) produces an immutable arithmetic sequence.

# Behaves like a read-only sequence:
# - index:    r[i]
# - slicing:  r[a:b]
# - len(r)
# - membership: x in r
# No interesting methods beyond what sequence protocol gives.


# -----------------------------------------------------------------------------
# 8. FILE OBJECT METHODS (FROM open())
# -----------------------------------------------------------------------------
# f = open("file.txt", "r", encoding="utf-8")
# Use with context manager:

# with open("file.txt") as f:
#     ...

# Read
# ----
# f.read()           → entire file as string
# f.readline()       → next line as string
# f.readlines()      → list of all lines

# Write
# -----
# f.write(s)         → write string, return number of chars
# f.writelines(list_of_strings)

# Position / control
# ------------------
# f.seek(offset, whence=0)
# f.tell()          → current byte position
# f.flush()
# f.close()


# -----------------------------------------------------------------------------
# 9. IMPORTANT MODULE FUNCTIONS (NOT BUILTINS)
# -----------------------------------------------------------------------------
# These are not global; they live in modules.

# math module
# -----------
# import math
# math.sqrt(x)
# math.floor(x)
# math.ceil(x)
# math.pow(x, y)
# math.pi, math.e

# random module
# -------------
# import random
# random.random()           → float in [0.0, 1.0)
# random.randint(a, b)      → int in [a, b]
# random.choice(seq)
# random.shuffle(list_in_place)

# statistics module (CS50P sometimes)
# -----------------------------------
# import statistics
# statistics.mean(data)
# statistics.median(data)
# statistics.mode(data)

# csv module (for file-based data processing)
# -------------------------------------------
# import csv
# csv.reader(file_obj)
# csv.writer(file_obj)
# Next level: DictReader, DictWriter

# json module
# -----------
# import json
# json.loads(s)    → parse JSON string to Python object
# json.dumps(obj)  → serialize Python object to JSON string


# -----------------------------------------------------------------------------
# 10. HIGH-LEVEL THINKING: "WHERE DOES THIS LIVE?"
# -----------------------------------------------------------------------------
# When you see a function/method in code, classify it:
#
# 1) Builtin?
#    - len, print, input, int, list, dict, sum, min, max, sorted, etc.
#
# 2) Method on an object?
#    - str methods: s.replace(), s.split(), s.lower()
#    - list methods: lst.append(), lst.sort(), lst.pop()
#    - dict methods: d.get(), d.items(), d.update()
#    - set methods: s.add(), s.union(), ...
#
# 3) Function from a module?
#    - math.sqrt(), random.randint(), json.loads(), csv.reader(), ...
#
# Once your brain auto-tags things like:
#   "replace → str method"
#   "append → list method"
#   "keys/items → dict methods"
#   "randint → random module"
# You’re officially thinking in Python’s object model, not just in “bag of tricks”.
# -----------------------------------------------------------------------------


# End of cheat sheet.
```

```python
# memoization- the property of a recursive fn to remember past returns of same fn arg called. So tht if a fn is called with same arg, it used the stored value here. (which will happen if a recur call of like fibbonacci)

dict_mem = {0:0, 1:1, 2:1}  # fibbonacci stars from pos 0
def fibbonacci(n):
    if n in dict_mem:
        return dict_mem[n]
    else:
        """         This is not req as added to dict instead
        if n == 1:
            return 0
        if n <= 3:
            return 1
        """
        return fibbonacci(n-1) + fibbonacci(n-2)
    
print(fibbonacci(7))
```

```python
# filter applies a fn to a seq, it cmps the returned value, and returns all inputs that returned true whn the fn was applied
def is_adult(age):
    if age >= 18:
        return True
    else:
        return False
    
ages = [34, 55, 12, 13, 44]
print(list( filter(is_adult, ages) ))   # 1 line
# output skips 12 and 13

# equivalent
output = []
for age in ages:                # 3 lines + slower
    if is_adult(age):
        output.append(age)

```

```python
# filter program, using lamda: ( no fn def )
ages = [34, 55, 12, 13, 44]         # 1 line + 1 line fn def
print(list( filter(lambda x: x>=17, ages) ))
# equivalent
ages = [ i for i in ages if ages >= 18 ]    # cond added at end of list comprehension

# storing the output, (instead of ret input after chking output by filter) done by fn map()
```

```python
# why use map and filter, when u can run a for loop and call a fn, instead of use fn inside map's argument?
    # it accelerates execution by reducing overhead of calling a fn each time- a fn call creates fn prefix state (setup) and return setsup main (postfix fn work).
        # this overhead of fn setup and restore is neglected by using a sequence on this fn and saving state until end of seq.
```

```python
# reduce- it iterates calling fn over a list. And each returned is used as arg for next ele in iterable
    # add(((a, b), c), d)
import functools
list_1 = range(1, 5)
print( functools.reduce(lambda x,y: x+y, list_1) )

# equivalent
def sum(a, b):
    return a + b
print( functools.reduce(sum, list_1))
```
