```python
# a 1D list stores reference of objs, when list_1 is appended, the whole ref of list_1 is copied to parent list_2.
    # On changing any elements of list_1, it will also reflect in the 2D list. appending ele to list_1 will show new ele in 2D parent list as well.
list_1 = [1, 2, 3, 4, 5]
parent_list = []
parent_list.append(list_1)  # parent list is a list of ptrs
list_1.append(7)
print(parent_list)
# will output parent list containing even the new changes to list_1

parent_list.append(list_1)  # creating a new ptr of list_1 to parent (also points to same obj list_1)
list_1.extend([6, 8])
print(parent_list)
# will output the new list twice, including latest changes
```

```python
# using fns in sets like common (intersection) and add both without duplicates (union)
s1 = {1, 3, 5}
s2 = {11, 3, 15}

s3 = s1 & s2    # intersection

s4 = s1 | s2    # union

print(f"Intersection of:\n{s1} and {s2} is: {s3}")
print(f"Union of:\n{s1} and {s2} is: {s4}")

# |||ly equivalent fns:

s3 = s1.intersection(s2)

s4 = s1.union(s3)

print(f"Intersection of:\n{s1} and {s2} is: {s3}")
print(f"Union of:\n{s1} and {s2} is: {s4}")

s5 = s1 - s2    # differnce
s5 = s1.difference(s2) # equivalent
print(f"Difference is: {s5}")

# unique terms of pool(s1 and s2)
s6 = s1.symmetric_difference(s2)
print(f"Symmertric difference is: {s6}")
s7 = (s1 | s2) - (s1 & s2)
print(f"Symmertric difference is: {s7}")    # equivalent, but diff ordering
```

```python
# zip
    # it is called a lazy iterator- it combies 2 elements into a single tuple, each tuple into a zip object
        # lazy iterator- it only stores references to the iterators. It does NOT store the zipped results; but returns a single tuple for each iteration, when iterating over a zip obj.
            #  It produces tuples by calling next() on each iterator, A one-shot, lazy iterator wrapper.
                # NOTE: if you iterate over a zip obj once, it will return None when u iterate over it again.
        # very fast and useful method to merge 2 sequences of attribute data, to one node with 2 elements.
        # note: can pass n args, returns obj with nodes; where each node is a tuple with n elements.
list_1 = list(range(10, 60, 10))
list_string = ["ten", "twenty", "thirty"]
obj_zip = zip(list_1, list_string)

for node in obj_zip:
    print(node)
```

```python
# if only 1 arguemnts is passed to a zip fn, it returns an obj with tuple nodes, where each tuple node has only 1 element
list_1 = list(range(10, 60, 10))
obj_zip = zip(list_1)
for node in obj_zip:
    print(node)
```

```python
# a program that has a menu of a cafe, a interactive CLI program
    # 2 fns '>' search '<' input

Menu_dict = {}
while True:
    print("Menu is:")
    for key in Menu_dict:
        print(f"{key}: {Menu_dict[key]}")

    print("$", end = '')
    line = input()
    print(line)
    if line[0] == 'q':
        break
    
    
    operator = line[0]
    line = line.replace(operator, "")
    line = line.split(':')
    key = line[0].strip()

    if operator == '>':
        if key in Menu_dict:
            print(Menu_dict[key])
        else:
            print(f"{key} not available in menu")

    elif operator == '<':
        value = line[1]
        Menu_dict[key] = value
        print(Menu_dict[key])

    else:
        print("Invalid operator")
    print()
    
```
