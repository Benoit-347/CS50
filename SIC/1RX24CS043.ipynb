```python
def factorial(n):
    # returns the factorial of the single number argmuent passed. (n! = n * n-1 * n-2)
    if n < 0:
        raise ValueError("Negative numbers not allowed")    # raise raises an error to python program, iterupting execution,  by terminating if not caught by except block
    result = 1
    for i in range(2, n+1): # multiples the number from 2 (as 1*n is n) until n times (n+1 is skipped; range property)
        result *= i
    return result

if __name__ == '__main__':
    n = int(input("Enter n: "))
    print("Factorial:", factorial(n))
```

```python
def fibonacci(n):
    # returns a list of all number of the fibbonacci seq until nth position of sequence, 0th is 0 and 1st is 1
    seq = []
    a, b = 0, 1
    while len(seq) < n:
        seq.append(a)
        a, b = b, a + b     # follows patter last_term + cur_term = next_term.
                                #The statement on left uses multi var assignment, feature of python
    return seq

if __name__ == '__main__':
    n = int(input("How many terms? "))
    print(fibonacci(n))
```

```python
def is_prime(n):
    # retuns boolen- True if prime no., False if not a prime no.
    if n < 2: return False  # a prime number is not -ve, its a natural number and 1 not prime
    i = 2
    while i*i <= n:     # a test to chk if prime as if numbers 1 upto sqrt(n) dont divide n with 0 remainder, then any number greater than sqrt(n) must be multiple by i, whr i< sqrt(n)
                            # a*b <= n; hence, for all a < sqrt(n), if n%a != 0, n is prime
        if n % i == 0:          # '%' means remainder
            return False
        i += 1
    return True

if __name__ == '__main__':
    n = int(input("Enter n: "))
    print("Prime" if is_prime(n) else "Not prime")
```

```python
def power(a, n):
    # returns the result of a**n with a recursion, ending recursion when n = 0.
    if n == 0:
        return 1
    if n < 0:
        return 1 / power(a, -n)
    return a * power(a, n-1)

if __name__ == '__main__':
    print(power(float(input("a: ")), int(input("n: "))))
```

```python
def is_pal(s):
    """checks if string is palindrome, by using additional fn, which calls itself, each time cmp left end of string to eight end of string
            if they dont match returns False, which ultimately returns False to parent fn. If it does not go False reaches the point where L > R, it return True to parent fn."""
    s2 = ''.join(ch.lower() for ch in s if ch.isalnum())
    def helper(l, r):
        if l >= r: return True
        if s2[l] != s2[r]: return False
        return helper(l+1, r-1)
    return helper(0, len(s2)-1)

if __name__ == '__main__':
    print(is_pal(input("String: ")))
```

```python
def count_digits(n):
    # returns the number of digits using a recur to '//' by 10, to shift left digits. until n is less than 10
    n = abs(n)
    if n < 10:
        return 1
    return 1 + count_digits(n//10)

if __name__ == '__main__':
    print(count_digits(int(input("n: "))))
```

```python
def gcd(a, b):
    """finds the highest divisor between a and b, using eclids algo,
        where the remainder of a division becomes the divisor, with the OLD divisor as the divident"""
    if b == 0:
        return abs(a)
    return gcd(b, a % b)

if __name__ == '__main__':
    a = int(input("a: ")); b = int(input("b: "))
    print(gcd(a,b))
```

```python
def recursive_max(lst):
    """uses recursion to obtain the largest element, storing it to l[0], base end of recur being when list len is 1
        and condition is checked if the the returned value is less than the first element of list """
    if len(lst) == 1:
        return lst[0]
    m = recursive_max(lst[1:])
    return lst[0] if lst[0] > m else m

if __name__ == '__main__':
    data = list(map(int, input("Numbers: ").split()))
    print(recursive_max(data))
```

```python
def binary_search(arr, low, high, target):
    # uses the target array, arr, the lower element and higher element, with target to search, condition grows logarithmically, as search in each recursion call changes search range to [mid, high] or [low, mid] where very iteration mid = (low + high) /2
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search(arr, mid+1, high, target)
    else:
        return binary_search(arr, low, mid-1, target)

if __name__ == '__main__':
    arr = list(map(int, input("Sorted list: ").split()))
    t = int(input("Target: "))
    idx = binary_search(arr, 0, len(arr)-1, t)
    print("Found at index", idx if idx!=-1 else "Not found")
```

```python
def linear_search(lst, i, target):
    # linear seach is done by linearly calling fn which check 1 pos each call 'i', where i determines the index the fn will search for target.
    if i >= len(lst):
        return -1
    if lst[i] == target:
        return i
    return linear_search(lst, i+1, target)

if __name__ == '__main__':
    lst = input("List: ").split()
    print(linear_search(lst, 0, input("Target: ")))
```

```python
def sum_list(lst):
    """recursive add the return of previous return and curent list element,
        starting from the first element, calling until the end of the list is reached, a non list,
        and returns the value to each parent call, until it end whn it adds with first element."""
    if not lst:
        return 0
    return lst[0] + sum_list(lst[1:])

if __name__ == '__main__':
    data = list(map(int, input("Nums: ").split()))
    print(sum_list(data))
```

```python
def hanoi(n, source, target, aux):
    # uses recursion to move disk to target from source, as base case (n ==1); If n not 1, move disc from source to aux recursively, and on returning, moves disc from aux to target recursively.
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    hanoi(n-1, source, aux, target)
    print(f"Move disk {n} from {source} to {target}")
    hanoi(n-1, aux, target, source)

if __name__ == '__main__':
    n = int(input("Disks: "))
    hanoi(n, 'A', 'C', 'B')
```

```python
def second_largest(arr):
    """returns the second largest number by  using helper fn, when returns the last element if, the passed index on heper is last index
       else: starts from index 0 recursively upto last index, returning largest and second largest element respectively """
    if len(arr) < 2:
        return None
    def helper(index):
        if index == len(arr)-1:
            return (arr[index], None)  # (max, second)
        m, s = helper(index+1)
        v = arr[index]
        if s is None:
            if v > m: return (v, m)
            elif v == m: return (m, None)
            else: return (m, v)
        # both exist
        if v > m: return (v, m)
        elif m > v > s: return (m, v)
        else: return (m, s)
    _, sec = helper(0)
    return sec

if __name__ == '__main__':
    a = list(map(int, input("List: ").split()))
    print("Second largest:", second_largest(a))
```

```python
def permutations(s):
    """ finds all combinations of string """
    if len(s) <= 1:
        return [s]
    res = []
    for i, ch in enumerate(s):
        for p in permutations(s[:i] + s[i+1:]):
            res.append(ch + p)
    return res

if __name__ == '__main__':
    print(permutations(input("String: ")))

```

```python
def make_adder(x):
    # hsa a closure variable x, which is stored at firstcall, then onl calling again, adds prev arg x, to the called new arg y; returning x + y
    def adder(y):
        return x + y
    return adder

if __name__ == '__main__':
    add5 = make_adder(5)
    print(add5(10))
```

```python
def make_multiplier(n):
    # closure on x, rturning x*y
    def mul(x):
        return x * n
    return mul

if __name__ == '__main__':
    doubler = make_multiplier(2)
    print(doubler(7))
```

```python
def make_power(n):
    # closure on n, rturning n**y
    def power(x):
        return x ** n
    return power

if __name__ == '__main__':
    square = make_power(2)
    print(square(5))
```

```python
def counter():
    # closure on varable counter, allocating it on first call, intializes it to 0, and every subsequent call, increments counter by 1.
    count = 0
    def inc():
        nonlocal count
        count += 1
        return count
    return inc

if __name__ == '__main__':
    c = counter()
    print(c(), c(), c())
```

```python
def call_tracker(func):
    """has parameter 'func' which is another fn passed to tracker, calls tracker to initialize wrapper fn,
        every subsequent call of 'func' will print count values and call 'func' """
    count = 0
    def wrapper(*args, **kwargs):
        nonlocal count
        count += 1
        print(f"Call #{count} to {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

if __name__ == '__main__':
    @call_tracker   # calls call_tracker on next fn
    def greet(name): print("Helo", name)
    greet("Alice"); greet("Bob")
```

```python
def prefixer(prefix):
    # assigns a prefix on first call, every subsequent call prints prefix along with the new string in arg
    def add_prefix(msg):
        return f"{prefix} {msg}"
    return add_prefix

if __name__ == '__main__':
    p = prefixer("[INFO]")
    print(p("System running"))
```

```python
def greeter(name):
    # assigns a name to closure, every subsequent call will print 'hello name'
    def greet():
        return f"Hello, {name}!"
    return greet

if __name__ == '__main__':
    say = greeter("Ravi")
    print(say())
```

```python
def password_validator(min_len=8):
    """assign min length of password on first call,
           every subsequent call checks if new arg is less than closure len,
               and if at least one number exists inside string."""
    def validate(pw):
        return len(pw) >= min_len and any(ch.isdigit() for ch in pw)
    return validate

if __name__ == '__main__':
    v = password_validator(6)
    print(v("abc123"), v("short"))
```

```python
def running_total():
    """first call initialises var total to 0,
      every subsequent call add new arg x, to total."""
    total = 0
    def add(x):
        nonlocal total
        total += x
        return total
    return add

if __name__ == '__main__':
    rt = running_total()
    print(rt(5), rt(10), rt(-3))
```

```python
def area_with_pi(pi):
    """to calculate area, first call initialises pi value, every subsequent call will multiply r**2 * pi;
          Returning the area with out own pi value."""
    def area_circle(r):
        return pi * r * r
    return area_circle

if __name__ == '__main__':
    area_custom = area_with_pi(3.14159)
    print(area_custom(2))
```

```python
def calculator(x):
    """initialises the operand 1 in first call, and every subsequent call takes in 1 operator and 1 more operand,
           returning the eval of expression: operand_1 operator operand_2"""
    def operate(y, op):
        if op == '+': return x + y
        if op == '-': return x - y
        if op == '*': return x * y
        if op == '/': return x / y
        raise ValueError("Unknown op")
    return operate

if __name__ == '__main__':
    calc5 = calculator(5)
    print(calc5(3, '+'), calc5(2, '*'))
```

```python
def memoize_factorial():
    """first call initialises the cache variable of type dict,
       every subsequent call calculates factorial, by checking cache variable first,
       if value not found here, it calculates factorial and inserts the key (the arg) and calculated value (the result) into cache var
       and returns the value."""
    cache = {}
    def fact(n):
        if n in cache:
            return cache[n]
        if n == 0:
            res = 1
        else:
            res = n * fact(n-1)
        cache[n] = res
        return res
    return fact

if __name__ == '__main__':
    mf = memoize_factorial()
    print(mf(10))
```

```python
def memoize_fib():
    """first call initialises the cache variable of type dict,
       every subsequent call calculates fibonnacci, by checking cache variable first,
       if value not found here, it calculates fibinnacci and inserts the key (the arg) and calculated value (the result) into cache var
       and returns the value."""
    cache = {0:0,1:1}
    def fib(n):
        if n in cache:
            return cache[n]
        cache[n] = fib(n-1) + fib(n-2)
        return cache[n]
    return fib

if __name__ == '__main__':
    f = memoize_fib()
    print(f(30))
```

```python
def id_generator(prefix="ID"):
    """First call initialises variable count, every subsequent call returns a id,
           which will be a increment of count, hence also being unique. """
    count = 0
    def gen():
        nonlocal count
        count += 1
        return f"{prefix}{count:04d}"
    return gen

if __name__ == '__main__':
    gen = id_generator("USR")
    print(gen(), gen())
```

```python
def threshold_filter(th):
    """First call initialises max number (variable th)
           every subsequent call passes a list, among which only elements smaller than th is returned into a new list;
           using list comprehension."""
    def keep(lst):
        return [x for x in lst if x > th]
    return keep

if __name__ == '__main__':
    f = threshold_filter(10)
    print(f([5,12,7,20]))
```

```python
def last_n_recorder(n):
    """initialises a list 'buf' which acts like a buffer with max n items,
           every subsequent call, add the new arg to this list,
           if the number of elements in this buf exceeds n, remove the first element in buf, hence acting like a queue."""
    buf = []
    def record(x):
        buf.append(x)
        if len(buf) > n:
            buf.pop(0)
        return list(buf)
    return record

if __name__ == '__main__':
    rec = last_n_recorder(3)
    print(rec(1), rec(2), rec(3), rec(4))
```

```python
def make_replacer(old, new):
    """Initiates old and new variables, which
           on every subsequent call replaces every old substring in new arg (str) with new substring."""
    def replace(s):
        return s.replace(old, new)
    return replace

if __name__ == '__main__':
    rep = make_replacer("cat", "dog")
    print(rep("The cat sat on the cat."))
```

```python
def discounter(rate):
    """Initiates the variable rate,
       every subsequent call returns the new argument with rate % deducted from it."""
    def apply(price):
        return price * (1 - rate)
    return apply

if __name__ == '__main__':
    disc10 = discounter(0.1)
    print(disc10(250))
```

```python
def temp_converter(scale):
    """Initialises the scale variants to be either 'C' or 'F',
       Every subsequent call, depending on if scale is 'C' or 'F', converts new arg to F unit on temperature or C unit respectively"""
    def convert(c):
        if scale == 'F': return (c * 9/5) + 32
        if scale == 'K': return c + 273.15
        raise ValueError("Unknown scale")
    return convert

if __name__ == '__main__':
    toF = temp_converter('F')
    print(toF(37))
```

```python
def divisible_by(d):
    """First call initiaises divisor,
        every subsequent call checks if the new arg is divisible by this divisor"""
    def check(n):
        return n % d == 0
    return check

if __name__ == '__main__':
    div3 = divisible_by(3)
    print(div3(9), div3(10))
```

```python
def table_for(n):
    """First call Stores the number for which a multiplication table is to be made,
        subsequent calls, print the table from 1 to 'upto' times"""
    def table(upto=10):
        return [n*i for i in range(1, upto+1)]
    return table

if __name__ == '__main__':
    t5 = table_for(5)
    print(t5(10))
```

```python
def rect_perimeter_fixed_width(width):
    """Initialises the width of the rectanlge, every subsequent call recieves lenght, and it returns the perimiter of the rectanlge, using the parent fns, width"""
    def perimeter(length):
        return 2*(width + length)
    return perimeter

if __name__ == '__main__':
    p = rect_perimeter_fixed_width(3)
    print(p(10))
```

```python
def caesar_cipher(shift):
    """Initializes the 'shift' variable, which for every subsequent call, shits the character 'shift' times, circling around form 'z' to 'a', then 'b' and so on..."""
    def encrypt(s):
        res = []
        for ch in s:
            if ch.isalpha():
                base = ord('A') if ch.isupper() else ord('a')
                res.append(chr((ord(ch)-base+shift)%26 + base))
            else:
                res.append(ch)
        return ''.join(res)
    return encrypt

if __name__ == '__main__':
    enc = caesar_cipher(3)
    print(enc("Hello, World!"))
```

```python
import time
def make_timer():
    """Initialises the 'start' variable, and stores the current time to it, it is later used aon every subsequent call, to find the differnce between that old 'start' time and the time during new calls. """
    start = time.time()
    def elapsed():
        return time.time() - start
    return elapsed

if __name__ == '__main__':
    t = make_timer()
    input("Press Enter after some time...")
    print("Elapsed seconds:", round(t(), 3))
```
