```python
# using property declarator; functionlity of class changed such tht- now class is passed as arg
    # the property tht a class in passed into arg, requiring @classmethod property; is imp whn you want to use inheritance in classes, to create sub-classes.
class Student:

    @classmethod
    def class_print(cls, string):
        return f"<class 'Student'> {string}"

print(Student.class_print("hello world"))
```

```python
import random
# using class members, that will be visible even without a obj
class Student:

    list_houses = ["green", "yellow", "blue", "red"]

    @classmethod
    def class_print(cls, string):
        return f"<class 'Student'> {string}"

    # You may use direct access to class variables, or use- the property tht a class in passed into arg with @classmethod. 
        # the above property is very imp to use inheritance in classes, to create sub-classes.

    @classmethod
    def class_rand_print(cls):
        name = random.choice(cls.list_houses)
        return f"House name from <class 'Student'> {name}"


print(Student.class_rand_print())

# note: (as outside methods, accessed by class.var; but inside method, may use cls.var or class.var).
    # asscessing class vars outside a fn- (ez) class.var
print(Student.list_houses)
```

```python
# you can use a class to organize code of different functionalities/ diff people; use modules to orgaize fns and programs; or stick to local files. its a question of which is imp whn.

class Student:
    def __init__(self, name, house):
        self.name = name
        self.house = house

    @classmethod
    def get_student(cls):
        name = input("Enter name: ")
        house = input("Enter house: ")
        return cls(name, house)
    
obj = Student.get_student() # calling method from class itself to create an object  # where last time we used a method externally, outside the class
print(f"obj.name: {obj.name}, obj.house: {obj.house}")
```

```python
# using class directly instead of using properties:
class Student:
    def __init__(self, name, house):
        self.name = name
        self.house = house

    def get_student():
        name = input("Enter name: ")
        house = input("Enter house: ")
        return Student(name, house)
    
obj = Student.get_student() # calling method from class itself to create an object  # where last time we used a method externally, outside the class
print(f"obj.name: {obj.name}, obj.house: {obj.house}")
```
