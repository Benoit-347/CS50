```python
class Fruits:
    def __init__(self, Orange = 0, Mango = 0, Apple = 0):
        Fruits.Orange = Orange
        Fruits.Mango = Mango
        Fruits.Apple = Apple
    def __str__(self):
        return f"Oranges: {self.Orange}, Mangoes: {self.Mango}, Apples: {self.Apple}"
person_1 = Fruits(5, 10, 15)
person_2 = Fruits(10, 20, 10)
print(person_1)
print(person_2)

orange_n = person_1.Orange + person_2.Orange
mango_n = person_1.Mango + person_2.Mango
apple_n = person_1.Apple + person_2.Apple

total = Fruits(orange_n, mango_n, apple_n)
print(total)

# what if we can use a single operator, to add all the attributs from 2 classes to return the result into another obj
```

```python
class Fruits:
    def __init__(self, Orange, Mango, Apple):
        self.Orange = Orange
        self.Mango = Mango
        self.Apple = Apple

    def __str__(self):
        return f"Orange: {self.Orange}, Mango: {self.Mango}, Apple: {self.Apple}"

    # __add__ init method allows using an operator overloading, to intepret it to be the fn body of __add__
    def __add__(self, self_2):  # LHS operand obj + RHS operand obj, uses the objs of both the operands next to '+'
        o = self.Orange + self_2.Orange
        m = self.Orange + self_2.Mango
        a = self.Apple + self_2.Apple
        return Fruits(o, m, a)
    
person_1 = Fruits(10, 20, 30)
person_2 = Fruits(20, 40, 60)
print(person_1)
print(person_2)
total = person_1 + person_2

print(f"Total is:\n{total}")
```
