```python
def main1():
    name, house = get_student()
    print(f"{name} is in {house} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return name, house  # We can only return 1 data item as a whole;
                            # hence this would be compiled as return (name, house)

# here: 
# return name, house -> is taken as tuple i.e:
# return (name, house)

main1()
```

```python
#using a tuple instead of py method

def main2():
    student = get_student() # tuple
    print(f"{student[0]} is in {student[1]} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return (name, house)  #We can only return 1 data item as a whole

main2()
```

```python
# a feature that if the name is padma, change house to Ravenclaw
def main3():
    student = get_student()
    if student[0] == "Harry":
        student[1] = " Gryffindor"
    print(f"{student[0]} is in {student[1]} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return (name, house)
main3()
```

```python
# also when we use list/tuples, we often can make mistakes with indexes as all access operation would use numbers;
    # so it may be better to use dicts (especially in large programs; also numbers are hard to remember)
def main4():
    student = get_student()
    for i in student:
        if i == "Harry":
            student[i] = "Gryffindor"
        print(f"{i} is in {student[i]} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return {name: house}
main4()
```

```python
# we still want to get key names by accessing storage, how to go around that? use the keys u want to store, as values to 3rd party key (name i.e- column_name/attribute), instead of the data itself
# this is standard as key names shoud be unique

def main5():
    student = get_student()
    if student["name"] == "Harry":
        student["house"] = "Gryffindor"
    print(f"{student['name']} is in {student['house']} house")

def get_student():
    name = input("Enter Name: ")
    house = input("Enter House: ")
    return {"name": name, "house": house}

main5()
```
